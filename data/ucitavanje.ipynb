```python
import os
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
directory = "facebook/facebook"  

all_feat_names = set()

for filename in os.listdir(directory):
    if filename.endswith(".featnames"):
        ego_id = int(filename.split('.')[0])
        feat_names_file = os.path.join(directory, f"{ego_id}.featnames")
        with open(feat_names_file, 'r') as file:
            feat_names = [line.strip() for line in file]
            all_feat_names.update(feat_names)

ordered_feat_names = sorted(all_feat_names)
```
